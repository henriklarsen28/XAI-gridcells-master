```python
import matplotlib.pyplot as plt
import json

path = 'goal_visitations.json'

with open(path, 'r') as f:
    data = json.load(f)

# Find common keys only
model_num = sorted(set(data["regular"].keys()))

# Extract y-values
regular_y = [data["regular"][k] for k in model_num]
rotated_y = [data["rotated"][k] for k in model_num]

# Plotting with shared keys only
plt.figure(figsize=(12, 6))
plt.plot(model_num, regular_y, marker='o', label='Regular two rooms environment (previous goal area)')
plt.plot(model_num, rotated_y, marker='x', label='Rotated two rooms environment (current goal area)')
plt.xlabel("Model number")
plt.ylabel("Number of visitations to goal area")
plt.title("Regular vs Rotated")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
